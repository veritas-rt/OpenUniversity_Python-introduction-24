# 1.2 文字列

##文字列

文字列は、文字が0個以上つながったものである。その文字の個数のことを「長さ」と呼ぶ。

*  文字列は、ダブルクォーテーションマークか、シングルクォーテーションマークのどちらかで、左右を挟んで表現する。どちらを使用してもよい。
 * 例 "Chiba"
 * 例 'Makuhari'
* "Wakaba' はダブルクォーテーションマークで始まり、シングルクォーテーションマークで終わっているため、文字列ではない。
* 長さ が 0 の文字列は、空列と呼ばれる。以下のどちらかで表される。
  * "" ダブルクォーテーションマーク2つ連続
  * '' シングルクォーテーションマーク2つ連続

## 変数としての文字列

数学では、「変数」というと、ひとつのアルファベットを使っていて、「数」を代入する、という前提で考えることが多いが、コンピュータのプログラムでは、「変数」には、様々な対象を代入することができる。また、変数名は、アルファベットや数字・記号を並べたものを利用することができる。そのため、変数名に英語や日本語（ローマ字）の意味のある名前を利用することがおおい。

最初に取り扱う文字列もまた、変数として取り扱うことができる。実際に、文字列を代入したあと、文字列の結合を行ってみよう。

【プログラム1203】

```
address1 = "Chiba"
address2 = 'Makuhari'
address3 = address1 + address2
print(address3)
```

##プログラム中にメモを書く、空行の取り扱い

プログラムの中にメモを書くことができる。メモは、プログラム実行時に、全く無視される。プログラム中に、＃　を記せば、そこから右側はメモとなる。

この実行されない部分のことを「コメント」といい、コメントになっている部分を「コメントアウトされた部分」と言う。

また、何も書いていない行は「空行」と呼ばれ、無視される。

【プログラム1204】

```
# 文字列の結合のプログラム

# まず、文字列を定義する。
address1 = "Chiba" # 千葉
address2 = 'Makuhari' # 幕張

# 結合させる
address3 = address1+ address2

# 表示する
print(address3)
```

## 文字列を挟む記号の使い方

文字列は、シングルクォーテーションマークか、あるいは、ダブルクォーテーションマークではさむことになっている。通常は、どちらを使用してもよいが、次の場合は例外となる。

+ シングルクォーテーションマークを文字列に含みたい場合は、その文字列をダブルクォーテーションマークで挟む。
+ ダブルクォーテーションマークを文字列に含みたい場合は、その文字列をシングルクォーテーションマークで挟む。

【プログラム1205】

```
# シングルクォーテーションマークを文字列に含みたい場合
a = "シングルクォーテーションマークは ' です。"
print(a)

# ダブルクォーテーションマークを文字列に含みたい場合
b = 'ダブルクォーテーションマークは " です。'
print(b)
```

## 改行させないで表示する方法

print 命令の最後に、end=""　をつけると、表示のあと、改行しない。

【プログラム1207】

```
# こちらは改行する
print("abc")
print("def")
```

【プログラム1208】

```
# こちらは改行しない
print("abc", end="")
print("def")
```

### 文字列に改行を入れてしまう方法

改行文字と呼ばれる文字があり、これを「表示」すると、強制改行になる。Pythonでの改行文字は、文字列中に \\n と書く。

* 文字 \\ は、ブラウザやパソコンによっては、￥の半角文字（円記号）で表示されることがあるが、同一のものである。（厳密には違うときもありますが……）
* Macで \ を入力するときは、option キー（左下）を押しながら￥を押す。

【プログラム1209】

```
# 改行文字を含む文字列を代入して表示
x = "higashi\nnishi"
print(x)
```

【プログラム1210】

```
# a, b は改行文字を含まない
a = "apple"
b = "banana"

# そのまま文字列を結合させるとこうなる
c = a + b
print(c)

# もし、d に改行文字を含めると、こうなる
d = a + "\n" + b
print(d)
```

### 注意（文字列の結合にカンマは使えない）

なお、文字列同士を カンマ で結合させることはできない。後述するタプルになってしまう。

【プログラム1211】

```
a1 = "Chiba"
a2 = "Makuhari"

# タプルを代入する
a3 = a1,a2
print(a3)
```

## 文字や文字列を繰り返して文字列を作る

Python では、同じ文字列が並んだ文字列を簡単に作ることができる。

【プログラム1212】

```
# アンダースコア _ を40回並べて区切り線を作る
kugiri = "_" * 40
print(kugiri)

# MW を 20回ならべて模様を作る
moyo = "MW" * 20
print(moyo)
```
